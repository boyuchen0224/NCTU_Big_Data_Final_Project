```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
df = pd.read_csv('./all_info_utf8.csv')
```

## List crops which trading day is larger than 2300

```python
market = np.unique(df['市場名稱'])
arr = np.unique(df.loc[df['市場名稱'] == '台北一']['作物名稱'])
arr2 = []
for crop in list(arr):
    try:
        taipei1 = df.loc[(df['作物名稱'] == str(crop)) & (df['市場名稱'] == '台北一')]
        if(len(taipei1) > 2300):
            print('taipei1', len(taipei1), 'crop', crop)
            arr2.append(str(crop))
    except:
        pass
```

## Plot the trading chart

```python
# arr = ['鳳梨-金鑽鳳梨', '香蕉', '青蔥-日蔥','蓮霧-紅蓮霧', '芽菜類-黃豆牙', '南瓜-木瓜形', '大蒜-蒜瓣', '小白菜-土白菜'\
#        '木瓜-網室紅肉', '洋香瓜-網狀紅肉']
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'svg'

for crop in arr2:
    df_tmp = df.loc[(df['作物名稱'] == str(crop)) & (df['市場名稱'] == '台北一')].sample(n = 1000)
    df_tmp = df_tmp.sort_values(by=['DateTime'])
    fig = px.line(df_tmp, x="DateTime", y="平均價", title='{} 歷年平均價格'.format(crop))
    fig.show(width=950, height=500)
```
