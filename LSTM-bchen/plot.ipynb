```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## Compare MSE between different batch_size

```python
# batch_size = [8, 16, 32, 64, 128, 256, 512]
batch_size = [64, 256, 512]
df_name = ['Price.csv', 'Price_fs.csv', '3_day_return.csv', '3_day_return_fs.csv']
color = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A', '#19D3F3', '#FF6692', '#B6E880']
crops_arr = [
    "香蕉",
    "洋香瓜-網狀紅肉",
    "蓮霧-紅蓮霧",
    "鳳梨-金鑽鳳梨",
    "南瓜-木瓜形",
    "大蒜-蒜仁",
    "青蔥-日蔥",
    "小白菜-土白菜",
    "芽菜類-黃豆牙",
    "木瓜-網室紅肉",
]

fig = make_subplots(
        rows=4, cols=1,
        subplot_titles=df_name,    
    )

for index, name in enumerate(df_name):
    for index2, batch in enumerate(batch_size):
        MSE = []
        for crops in crops_arr:
            df = pd.read_csv('./{}/{}_{}'.format(batch, crops, name))
            MSE.append(df['MSE'][0])
    
        fig.add_trace(
            go.Scatter(
                x=crops_arr,
                y=MSE,
                mode="lines+markers",
                name=batch,
                line=dict(color=color[index2])
            ),
            row=index + 1,
            col=1,
        )
fig.update_layout(
    autosize=False,
    width=1000,
    height=2000,
)
fig.show()
```
